```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
c=(2/np.sqrt(2*np.pi))*np.exp(1/2)
x=[]
i=0
while len(x)<10000:
    y=np.random.exponential(scale=1,size=1)
    u=np.random.uniform(low=0,high=1,size=1)
   #define threshold
    threshold=2*stats.norm.pdf(y,0,1)/(c*stats.expon.pdf(y))
    if u[0]<=threshold:
        if i%2==0:
            x.append(-1*y[0]) 
        else:
            x.append(y[0])
        i+=1
```

```python
real=np.random.normal(loc=0,scale=1,size=1000)
sns.distplot(x,hist=False,label='simulate',color='blue')
sns.distplot(real,hist=False,label='reality',color='red')
plt.legend(['simulate', 'reality'])
```

```python
g=np.random.standard_t(size=10000,df=3)
w=stats.norm.pdf(g,0,1)/stats.t.pdf(g,3)
np.mean(g*w)
```

```python
g=np.random.normal(size=10000,loc=0,scale=1)
w=stats.t.pdf(g,3)/stats.norm.pdf(g,0,1)
np.mean(g*w)
```
