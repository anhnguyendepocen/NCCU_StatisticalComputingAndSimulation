# define Gradient descent function

```python

##f=function
##df1=1st differential function
##x0=initial value
##rn=learning rate
##tol=tolerance
##maxiter=max iteration times
def Gradient(f,df1,x0,rn,tol=10**(-6),maxiter=50):
    xn=x0
    i=0
    while abs(df1(xn))>tol:
        xn=xn-rn*(df1(xn))
        i=i+1
        if i>maxiter:
            print('exceed max iteration times')
            break
      
  
    print("answer value=",xn,"iteration times=",i)
```

# define Newton function

```python
##df1=1st differential function
##df2=2nd differential function
##x0=initial value
##tol=tolerance
##maxiter=max iteration times
##findmin= want to fin min or not
def Newton(f,df1,df2,x0,tol=10**(-6),maxiter=50,findmin=True):
    xn=x0
    i=0
    if (findmin):
        while(abs(df1(xn))>tol or df2(xn)<0):
            xn=xn-(df1(xn)/df2(xn))
            i=i+1
            if(i>maxiter):
                print('exceed max iteration times')
                break
    else:
        while(abs(df1(xn))>tol):
            xn=xn-(df1(xn)/df2(xn))
            i=i+1
            if(i>maxiter):
                print('exceed max iteration times')
                break

    print("answer value=",xn,"iteration times=",i)
```

```python
def f(x):
    return x**3-2*x**2+2
def f_df1(x):
    return 3*x**2-4*x
def f_df2(x):
    return 6*x-4
```

```python
Gradient(f=f,df1 =f_df1,x0=2,rn=0.1)
```

```python
Gradient(f=f,df1 =f_df1,x0=0.01,rn=0.1)
```

```python
Newton(f=f,df1 =f_df1,df2=f_df2,x0=2,tol=10**(-6),maxiter=10000,findmin=True)
```

```python
Newton(f=f,df1 =f_df1,df2=f_df2,x0=0.01,tol=10**(-6),maxiter=10000,findmin=True)
```
