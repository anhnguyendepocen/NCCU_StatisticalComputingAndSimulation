```python
import numpy as np
from numpy.random import normal
import matplotlib.pyplot as plt
```

```python
i=np.arange(501)
ti=i/500
z=normal(loc=0,scale=1,size=501)
x=z*np.sqrt(1/500)
x[0]=0
w=np.cumsum(x)
```

```python
plt.plot(ti,w)
plt.xlabel('time')
plt.ylabel('value')
plt.title('simulate standard brownian motion')
```

```python
mu=5
sigma_sq=2
i=np.arange(501)
ti=i/500
z=normal(loc=0,scale=1,size=501)
x=z*np.sqrt(1/500)*np.sqrt(sigma_sq)+mu*(1/500)
x[1]=0
w=np.cumsum(x)
plt.plot(ti,w)
plt.xlabel('time')
plt.ylabel('value')
plt.title('simulate brownian motion')
```
