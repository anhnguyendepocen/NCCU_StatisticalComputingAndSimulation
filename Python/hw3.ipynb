```python
import numpy as np
from numpy.random import *
from scipy.stats import *
```

```python
means01=[]
for i in range(5000):
    g=normal(loc=0,scale=1,size=1000)
    w=uniform.pdf(g,-20,20)/norm.pdf(g,0,1)
    means01.append(np.mean(40*np.exp(-1*abs(g))*w))
print(np.mean(means01))
print(np.var(means01))
```

```python
means05=[]
for i in range(5000):
    g=normal(loc=0,scale=5,size=1000)
    w=uniform.pdf(g,-20,20)/norm.pdf(g,0,5)
    means05.append(np.mean(40*np.exp(-1*abs(g))*w))
print(np.mean(means05))
print(np.var(means05))
```

```python
meansmc=[]
for i in range(5000):
    f=np.random.uniform(low=-20,high=20,size=1000)
    meansmc.append(np.mean(40*np.exp((-1)*abs(f))))
print(np.mean(meansmc))
print(np.var(meansmc))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))
box=plt.boxplot([means01,means05,meansmc],patch_artist=True,labels=['sample N(0,1)','sample N(0,5)','MC method'])
box['boxes'][0].set(facecolo ="orange" )
box['boxes'][1].set(facecolor="red" )
box['boxes'][2].set(facecolor='blue')
plt.show()
```

```python
#define indicator function
def I_x3(x):
    out=[]
    for i in range(len(x)):
        if (x[i]>3):
            out.append(1)
        else :
            out.append(0)
    return out
```

```python
means2MC=[]
for i in range(10**5):
    f=normal(loc=0,scale=1,size=100)
    means2MC.append(np.mean(I_x3(f)))

print(np.mean(means2MC))
print(np.var(means2MC))
```

```python
means2IS=[]
for i in range(10**5):
    g=normal(loc=4,scale=1,size=100)
    w=norm.pdf(g,0,1)/norm.pdf(g,4,1)
    means2IS.append(np.mean(w*I_x3(g)))

print(np.mean(means2IS))
print(np.var(means2IS))
```

# compare variance

```python
np.var(means2MC)>np.var(means2IS)
```
