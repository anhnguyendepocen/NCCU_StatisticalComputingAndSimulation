# use non Monte Carlo

```python
import numpy as np
from numpy.random import normal
from scipy.stats import *
```

```python
N=5**5
v=10**5
a=[-3,-1,1,3,5]
hf=[]


for i1 in range(5):
    for i2 in range(5):
        for i3 in range(5):
            for i4 in range(5):
                for i5 in range(5):
                    hf.append(((a[i1]*a[i2]*a[i3]*a[i4]*a[i5])**2)*(norm.pdf(a[i1])*norm.pdf(a[i2])*norm.pdf(a[i3])*norm.pdf(a[i4])*norm.pdf(a[i5])))
I_1=v*np.mean(hf)
```

# use Monte Carlo

```python
N=5**5
h=[]
for i in range(N):
    h.append((normal(loc=0,scale=1,size=1)*normal(loc=0,scale=1,size=1)*normal(loc=0,scale=1,size=1)*normal(loc=0,scale=1,size=1)*normal(loc=0,scale=1,size=1))**2)
I_2=np.mean(h)
I_2
```
